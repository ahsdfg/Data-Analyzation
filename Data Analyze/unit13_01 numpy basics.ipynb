```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
# sin(a)의 평행이동
plt.plot(a + np.pi/2, np.sin(a))
plt.show()
```

```python
import numpy as np
a = np.arange(-5, 5)
print(a)
print(a<0)
# 마스크를 배열에 적용
print(a[a<0])

# 마스크를 변수에 넣어 활용
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0
print(a[mask1+mask2]) # 둘 중 하나의 조건이 참일 경우
print(a[mask1*mask2]) # 두 조건 모두 참일 경우
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 10~100 범위에서 200개 뽑기
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1+mask2]
y = y[mask1+mask2]
size = []
for i in range(len(x)):
    size.append(np.random.rand() * 100)
print(len(x), len(size))

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
