```python
# 인창동 인구수 시각화

import csv
import matplotlib.pyplot as plt

f = open('age.csv', encoding = 'utf8')
data = csv.reader(f)
result = []

for row in data:
    if '인창동' in row[0]:
        for i in row[3:]:
            result.append(int(i))

plt.rc('font', family='AppleGothic')
plt.title('인창동 인구수 시각화')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
f = open('age.csv', encoding = 'utf8')
data = csv.reader(f)
result = []
name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요: ")

for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))

plt.rc('font', family='AppleGothic')
plt.title(name + ' 지역의 인구 구조')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
# 인구구조 막대 그래프 

f = open('age.csv', encoding = 'utf8')
data = csv.reader(f)
result = []
town_name = input("인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요: ")

for row in data:
    if town_name in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
plt.bar(range(101), result)
plt.show()
```
