```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

name = input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i].replace(',', '')))
            f.append(int(row[i+103].replace(',', '')))
        break
        
plt.plot(m, label='male')
plt.plot(f, label='female')
plt.legend()
plt.show()
```

```python
f = open('gender.csv')
data = csv.reader(f)
result = []

name = input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            result.append(int(row[i].replace(',', '')) - int(row[i+103].replace(',', '')))
        break
        
plt.bar(range(101), result)
plt.show()
```

```python
# s = size, c = list 값으로 원하는 색상 선택해도 되고, range 값으로 주면 알아서 그 숫자만큼 정해줌
# cmap 컬러바에 사용될 색상의 종류 선택 가능
plt.scatter([1, 2, 3, 4], [10, 30, 20, 40], s=[30, 60, 90, 120], c=range(4), cmap='jet')
plt.colorbar()
plt.show()
```

```python
import random

x = []
y = []
size = []
for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))
# alpha = 투명도, 겹쳐도 보이게해줌
plt.scatter(x, y, s=size, c = x, cmap = 'jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []
for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))
plt.style.use('ggplot')
plt.scatter(x, y, s=size, c=size, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
import math
import csv

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
size = []

name = input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i].replace(',', '')))
            f.append(int(row[i+103].replace(',', '')))
            size.append(math.sqrt(int(row[i].replace(',', '')) + int(row[i+103].replace(',', ''))))
        break
        
plt.rc('font', family='AppleGothic')
plt.title(name + ' 지역의 성별 인구 그래프')
plt.style.use('ggplot')
plt.scatter(m, f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g') # 추세선 (남여비율이 같을 때)
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
