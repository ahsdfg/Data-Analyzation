```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
size = []

name = input('찾고 싶은 지역의 이름을 알려주세요: ')
for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(-int(i.replace(',', '')))
        for i in row[106:]:
            f.append(int(i.replace(',', '')))
        break # '제주'와 같이 여러 항목이 겹치는 상황 대비
            
            
size.append(-sum(m))
size.append(sum(f))
print(size)

plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi=300)
plt.barh(range(101), m, label='male')
plt.barh(range(101), f, label='female')
plt.legend()
plt.show()
```

```python
f = open('gender.csv')
data = csv.reader(f)
size = []


name = input('찾고 싶은 지역의 이름을 알려주세요: ')

for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3].replace(',', ''))
            f += int(row[i+106].replace(',', ''))
        break # '제주'와 같이 여러 항목이 겹치는 상황 대비
            
size.append(m)
size.append(f)

color = ['crimson', 'darkcyan']
plt.axis('equal')
# startangle = 시작각도, 반시계방향으로 90도 이동
plt.pie(size, labels=['m', 'f'], autopct='%.1f%%', colors=color, startangle=90)
plt.show()
```
